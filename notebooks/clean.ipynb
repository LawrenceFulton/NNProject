````python
import pandas as pd
import numpy as np
````

### Data import

````python
df = pd.read_csv("../data/raw_data/KNMI_20181231.csv", low_memory=False)
````

### Initial cleaning
+ YYYYMMDD to date/time object
+ Rename Columns to more descriptive names
    - (more to be done)

````python
df['YYYYMMDD'] = pd.to_datetime(df['YYYYMMDD'], format='%Y%m%d')
df = df.rename(columns={"STN"       : "Station", 
                        "YYYYMMDD"  : "Date", 
                        "DDVEC"     : "Wind_Direction_Vector",
                        "FHVEC"     : "Wind_Speed_Vector"})
df.head()
````

````python
df1, df2, df3, df4 = np.array_split(df, 4)
````

````python
df1.to_csv("../data/clean_data/df1.csv", index=False)
df2.to_csv("../data/clean_data/df2.csv", index=False)
df3.to_csv("../data/clean_data/df3.csv", index=False)
df4.to_csv("../data/clean_data/df4.csv", index=False)
````
